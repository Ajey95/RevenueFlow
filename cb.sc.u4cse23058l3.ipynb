```python
import re

text = """Your sample text goes here. You can paste #Vabber95, Vabniverse@ben.com, 6302101169,
URLs like https://ben10.com, or dates like 12/05/2023, or MM-DD-YYYY formats like 11-09-2022.Humming bird!"""
```

```python
# 1. Extract all email addresses
emails = re.findall(r'\b[\w.-]+@[\w.-]+\.\w+\b', text)
print("Emails:", emails)
```

```python
# 2. Find all dates in DD/MM/YYYY or MM-DD-YYYY format
dates = re.findall(r'\b(?:\d{2}/\d{2}/\d{4}|\d{2}-\d{2}-\d{4})\b', text)
print("Dates:", dates)
```

```python
# 3. Remove all punctuation
no_punct = re.sub(r'[^\w\s]', '', text)
print("No Punctuation:", no_punct)
```

```python
# 4. Extract hashtags
hashtags = re.findall(r'#\w+', text)
print("Hashtags:", hashtags)
```

```python
# 5. Validate Indian mobile numbers (starting with 6-9, 10 digits)
indian_numbers = re.findall(r'\b[6-9]\d{9}\b', text)
print("Indian Numbers:", indian_numbers)
```

```python
# 6. Find all words starting with a capital letter
capital_words = re.findall(r'\b[A-Z][a-zA-Z]*\b', text)
print("Capital Words:", capital_words)
```

```python
# 7. Replace all digits with #
replaced_digits = re.sub(r'\d', '#', text)
print("Replaced Digits:", replaced_digits)
```

```python
# 8. Split a paragraph into sentences
sentences = re.split(r'(?<=[.!?])\s+', text.strip())
print("Sentences:", sentences)
```

```python
# 9. Extract all URLs
urls = re.findall(r'https?://[^\s]+', text)
print("URLs:", urls)
```

```python
# 10. Find all words that end with "ing"
ing_words = re.findall(r'\b\w+ing\b', text)
print("Words ending with 'ing':", ing_words)
```
